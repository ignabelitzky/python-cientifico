# Transformada Discreta de Fourier

La transformada discreta de Fourier permite realizar el analisis espectral de una senal. Nos lleva del espacio de tiempo discreto al espacio de frecuencias, donde podemos obtener informacion sobre las componentes frecuenciales predominantes de una senal. El desarrollo de este modulo tendra el siguiente orden:

1. Definicion matematica de la transformada discreta de Fourier - Ejemplo en pequena senal
2. Identificacion frecuencia predominante en senal discreta
3. Identificacion de notas musicales con audio real
4. Analisis espectral en tiempo real usando entrada de microfono

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fourier
import scipy.io.wavfile as waves
```

## Ejemplo en pequena senal

```python
gn = [0, 1, 2, 3, 4] # Definimos una funcion en tiempo discreto
gk = fourier.fft(gn) # Calculamos la FFT
gk
```

## Identificacion Frecuencia Predominante en Senal Discreta

Crearemos una senal compuesta por dos ondas sinusoidales de 60 Hz. y 223 Hz. Agregaremos ruido a la senal compuesta y luego mediante la FFT identificaremso las frecuencias predominantes de la senal sintetica. 

```python
# Definimos un tiempo y frecuencia de muestreo
Ts = 0.001
Fs = 1.0 / Ts

w1 = 2 * np.pi * 60 # Definimos una frecuencia de 60 Hz para la senal 1
w2 = 2 * np.pi * 223 # Definimos una frecuencia de 223 Hz para la senal 2

t = Ts * np.arange(0, 1000)
ruido = np.random.random(len(t))
signal = 3 * np.sin(w1 * t) + 2.3 * np.sin(w2 * t) + ruido # Construimos una senal compuesta por la senal 1 y 2, y agregamos ruido

plt.xlim(0.2, 0.3)
plt.plot(t, signal, '.-')
plt.xlabel('Tiempo(s)', fontsize = '14')
plt.ylabel('Amplitud', fontsize = '14')
plt.show()
```

```python
gk = fourier.fft(signal)
M_gk = abs(gk)

for jj in range(0, 1000):
    if (M_gk[jj] > 400):
        print(jj, t[jj], M_gk[jj])
```

```python
F = Fs * np.arange(0, len(signal))/len(signal)

plt.plot(F, M_gk)
plt.xlabel("Frecuencia (Hz)", fontsize="14")
plt.ylabel("Amplitud FFT", fontsize="14")
plt.show()
```

## Identificacion de Notas Musicales con Audio Real

- Cargaremos una grabacion de audio .wav
- Obtendremos la variacion temporal de la senal de audio
- Encontraremos la frecuencia dominante y la compararemos con los rangos de frecuencias de las notas musicales

```python
# filename = 'rec_SOL.wav'
# filename = 'silb.wav'
filename = 'aaaa_short.wav'

Fs, data = waves.read(filename)
Audio_m = data[:, 0]

L = len(Audio_m)

t = np.arange(0, L)/Fs
plt.xlim(0.32, 0.34)
plt.plot(t, Audio_m)
plt.show()
```

```python
print(L, L//2)
gk = fourier.fft(Audio_m)
M_gk = abs(gk)
M_gk = M_gk[0:L//2]

Ph_gk = np.angle(gk)
F = Fs * np.arange(0, L//2)/L
plt.xlim(0, 5000)
plt.ylim(0, 6.0e7)
plt.plot(F, M_gk)
plt.xlabel("Frecuencia (Hz)", fontsize="14")
plt.ylabel("Amplitud FFT", fontsize="14")
plt.show()
```

```python
PosM = np.where(M_gk == np.max(M_gk))           # Encontramos la posicin para la cual la Magnitud de FFT es mxima
#Posm = np.where(M_gk == np.min(M_gk))           # Encontramos la posicin para la cual la Magnitud de FFT es mnima
F_fund = F[PosM]                                # Identificamos la Frecuencia asociada al valor del mximo de la Magnitud de FFT
#F_fund = F[Posm]                                # Identificamos la Frecuencia asociada al valor del mximo de la Magnitud de FFT
print(F_fund)
if F_fund > 135 and F_fund < 155:                   # Rango de frecuencias para nota RE
    print("La nota es RE, con frecuencia: ",F_fund)
elif F_fund > 190 and F_fund < 210:                   # Rango de frecuencias para nota SOL
    print("La nota es SOL, con frecuencia: ",F_fund)
elif F_fund > 235 and F_fund < 255:                     # Rango de frecuencias para nota SI
    print("La nota es SI, con frecuencia: ",F_fund)
elif F_fund > 320 and F_fund < 340:                   # Rango de frecuencias para nota MI
    print("La nota es MI, con frecuencia: ",F_fund)
```
