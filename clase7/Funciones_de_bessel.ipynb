## Funciones de Bessel de orden real (jv, jn_zeros)

Las funciones de Bessel son una familia de soluciones para la ecuacion diferencial de Bessel con alfa de orden real o complejo:
$$ x^{2} \frac{d^{2}y}{dx^{2}} + x \frac{dy}{dx} + (x^{2} - \alpha^{2})y = 0 $$
Entre otros usos, estas funciones surgen en problemas de propagacion de ondas, como los modos de vibracion de un cabezal de tambor delgado.

Una gran cantidad de los problemas mas diversos relacionados con casi todas las ramas mas importantes de la fisica matematica y desenados para responder preguntas tecnicas topicas estan asociados con la aplicacion de las ```funciones de Bessel.```

Las funcions de Bessel se usan ampliamente para resolver problemas de acustica, radiofisica, hidrodinamica, problemas de fisica atomica y nuclear. Numerosas aplicacions de las funciones de Bessel a la teoria de la ```condictividad termica y la teoria de la elasticidad (problemas en las vibraciones de la placa, problemas de la teoria de la cascara, problemas para determinar la concentracion de tension cerca de grietas).```

Tal popularidad de las funciones de Bessel se explica por el hecho de que resolver ecuaciones de fisica matematica que contienen el ```operador de Laplace en coordenadas cilindricas``` mediante el metodo clasico de separacion de variables conduce a una ecuacion diferencial ordinaria, que sirve para determinar estas funciones.

\begin{equation}
    \nabla^{2}\varPsi = \frac{\partial^{2}\varPsi}{\partial x^{2}} + \frac{\partial^{2}\varPsi}{\partial y^{2}} + \frac{\partial^{2} \varPsi}{\partial z^{2}} 
\end{equation}

\begin{equation}
    \nabla^{2}\varPsi = \frac{1}{\rho} \frac{\partial}{\partial \rho} \left(\rho \frac{\partial \varPsi}{\partial \rho}\right) \frac{1}{\rho^{2}} \frac{\partial^{2}\varPsi}{\partial\varphi^{2}} + \frac{\partial^{2} \varPsi}{\partial z^{2}}
\end{equation}

Si nos concentramos en el movimiento $\rho$, sin considerar $\varphi$ y $z$, nos queda

\begin{equation}
    \nabla^{2} \varPsi = \frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial\varPsi}{\partial \rho}\right)
\end{equation}

\begin{equation}
    \nabla^{2} \varPsi = \frac{1}{\rho} \left( \frac{\partial \varPsi}{\partial \rho} \right) + \left(\frac{\partial^{2} \varPsi}{\partial \rho^{2}} \right)
\end{equation}

### Funcion de Bessel de orden cero, uno y demas

```Scipy``` dispone de un paquete muy interesante llamado special donde se encuentran funciones como ```airy``` o ```bessel```.

[http://docs.scipy.org/doc/scipy/reference/tutorial/special.html](http://docs.scipy.org/doc/scipy/reference/tutorial/special.html)

La ecuacion diferencial de Bessel viene definida por,

\begin{equation}
    x^{2} \frac{d^{2} \varPsi}{dx^{2}} + x \frac{d\varPsi}{dx} + (x - \alpha) \varPsi = 0
\end{equation}

si tuvieramos $\alpha = n$ con $n \in \mathbb{N}$

\begin{equation}
    x^{2} \frac{d^{2} \varPsi}{d x^{2}} + x \frac{d \varPsi}{d x} + (x - n) \varPsi = 0
\end{equation}

cuya solucion es

\begin{equation}
    J_{n} (x) = \sum_{k = 0}^{\infty} \frac{(-1)^{k} (\frac{x}{2})^{n+2k}}{k!(n+k)!}
\end{equation}

para $n = 0$ y $n = 1$ obtenemos las ```funciones de Bessel de orden 0 y 1.```

Estas funciones son de gran importancia ya que las funciones del resto de ordenes se pueden escribir en funcion de estas dos.

\begin{equation}
    J_{0} (x) = 1 - \frac{(\frac{x}{2})^{2}}{(1!)^{2}} + \frac{(\frac{x}{2})^{4}}{(2!)^{2}} - \frac{(\frac{x}{2})^{6}}{(3!)^{2}} + \cdots
\end{equation}

\begin{equation}
    J_{1} (x) = \frac{x}{2} \Bigg[ 1 - \frac{(\frac{x}{2})^{2}}{(1!2!)^{2}} + \frac{(\frac{x}{2})^{4}}{(2!3!)^{2}} - \frac{(\frac{x}{2})^{6}}{(3!4!)^{2}} + \cdots \Bigg]
\end{equation}

Calculemos los coeficientes de orden cero, uno y demas de la funcion de Bessel contenida en el paquete ```special```.

````python
from scipy import special  # Importamos scipy.special
import scipy as sp
import numpy as np
````

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 3D graph type
from matplotlib import cm  # color maps
````

### Graficos de las funciones de Bessel

A continuacion observaremos la forma de estas funciones para distintos ordenes.

````python
# Funcion de Bessel de orden n
plt.figure(figsize=(10,7))
x = np.linspace(0, 20, 200)
n = 0  # Orden de la funcion de Bessel
texto = 'Bessel J' + str(n)
plt.plot(x, special.jn(n, x), 'b', label= texto) # special.jn(n, r * kth_zero)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title(texto)
plt.grid(True)
````

````python
# Creamos el array dimensional de espacio de las x
x = np.arange(0, 30, .1)

# Calculamos los coeficientes de orden cero.
j0 = special.j0(x)

# Calculamos los coeficientes de orden uno.
j1 = special.j1(x)

# Calculamos los coeficientes de orden dos.
j2 = special.jn(2, x)

#Calculamos los coeficientes de orden tres.
j3 = special.jn(3, x)
````

````python
plt.figure(figsize=(10,7))

# Dibujamos la funcion de Bessel de orden cero.
plt.plot(x, j0, 'b', label='Orden cero')

# Dibujamos la funcion de Bessel de orden uno.
plt.plot(x, j1, 'g', label='Orden uno')

# Dibujamos la funcion de Bessel de orden dos.
plt.plot(x, j2, 'r', label='Orden dos')

# Dibujamos la funcion de Bessel de orden tres.
plt.plot(x, j3, 'k', label='Orden tres')

# Anadimos la leyenda
plt.legend()
plt.xlabel('x')
plt.ylabel('J(x)')

# Mostramos la grafica
plt.grid(True)
plt.show()
````
